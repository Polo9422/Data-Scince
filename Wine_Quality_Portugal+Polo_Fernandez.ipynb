```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame llamado 'df' con tus datos de vinos

# Gráfico 1: Scatterplot (fixed acidity vs volatile acidity vs pH)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='fixed acidity', y='volatile acidity', hue='pH')
plt.title('Fixed Acidity vs Volatile Acidity vs pH')
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.grid(True)
plt.legend(title='pH')
plt.show()


```

```python
# Gráfico 3: Gráfico de línea (free sulfur dioxide vs total sulfur dioxide vs sulphates)
plt.figure(figsize=(10, 6))
df[['free sulfur dioxide', 'total sulfur dioxide', 'sulphates']].plot(kind='line')
plt.title('Free SO2, Total SO2, and Sulphates Over Time')
plt.xlabel('Observation')
plt.ylabel('Value')
plt.grid(True)
plt.legend(title='Variables')
plt.show()
```

```python
# Gráfico 2: Gráfico de barras (citric acid, residual sugar y alcohol)
plt.figure(figsize=(10, 6))
sns.barplot(data=df[['citric acid', 'residual sugar', 'alcohol']].mean(), orient='h')
plt.title('Average Citric Acid, Residual Sugar, and Alcohol Content')
plt.xlabel('Average Value')
plt.ylabel('Variables')
plt.show()
```
